New Customer Ratio

Initially, it's essential to establish the criteria for identifying a new customer. In our dataset, we'll consider a new customer as someone making their first purchase within the specified time window, and for this example, we'll analyze this on a monthly basis.

To identify new customers, we'll employ the .min() function to determine the first purchase date for each customer. The subsequent code will apply this function, enabling us to categorize customers as new based on their first purchase date. The resulting code will then showcase the monthly revenue breakdown for each customer group.

```python
#create a dataframe contaning CustomerID and first purchase date
tx_min_purchase = df.groupby('CustomerID').FormattedInvoiceDate.min().reset_index()
tx_min_purchase.columns = ['CustomerID','MinPurchaseDate']
tx_min_purchase.head()
```

```python
tx_min_purchase['MinPurchaseYearMonth'] = tx_min_purchase['MinPurchaseDate'].map(lambda date: 100 * date.year + date.month)
```

```python
#merge first purchase date column to our main dataframe (tx_uk)
df = pd.merge(df, tx_min_purchase, on='CustomerID')

df.head()
```

```python
#creating a column named User Type and assign Existing if User's First Purchase Year Month before the selected Invoice Year Month
df['UserType'] = 'New'
df.loc[df['InvoiceYearMonth']>df['MinPurchaseYearMonth'],'UserType'] = 'Existing'
```

```python
#calculate the Revenue per month for each user type
tx_user_type_revenue = df.groupby(['InvoiceYearMonth','UserType'])['Revenue'].sum().reset_index()
tx_user_type_revenue.head()
```

```python
tx_user_type_revenue = tx_user_type_revenue.query("InvoiceYearMonth != 201012 and InvoiceYearMonth != 201112")
tx_user_type_revenue.head()
```

```python
#filtering the dates and plot the result

plot_data = [
    go.Scatter(
        x=tx_user_type_revenue.query("UserType == 'Existing'")['InvoiceYearMonth'],
        y=tx_user_type_revenue.query("UserType == 'Existing'")['Revenue'],
        name = 'Existing'
    ),
    go.Scatter(
        x=tx_user_type_revenue.query("UserType == 'New'")['InvoiceYearMonth'],
        y=tx_user_type_revenue.query("UserType == 'New'")['Revenue'],
        name = 'New'
    )
]

plot_layout = go.Layout(
        xaxis={"type": "category"},
        title='New vs Existing'
    )
fig = go.Figure(data=plot_data, layout=plot_layout)
pyoff.iplot(fig)
```


Existing customer trends indicate a positive trajectory, suggesting growth in our customer base. However, new customers exhibit a marginal negative trend. To gain a better idea, let's examine the New Customer Ratio.

**Monthly Retention Rate**

Close attention should be paid to the retention rate as it reflects the stickiness of your service and the alignment of your product with the market. To visualize the Monthly Retention Rate, it's crucial to compute the number of customers retained from the previous month.

Monthly Retention Rate = Retained Customers From Prev. Month/Active Customers Total

Using crosstab() function of to calculate the Retention Rate.

```python
#identify which users are active by looking at their revenue per month
tx_user_purchase = df.groupby(['CustomerID','InvoiceYearMonth'])['Revenue'].sum().reset_index()
tx_user_purchase.head()
```

```python
#create retention matrix with crosstab
tx_retention = pd.crosstab(tx_user_purchase['CustomerID'], tx_user_purchase['InvoiceYearMonth']).reset_index()

tx_retention.head()
```

```python
# Creating an array of dictionaries to store Retained & Total User count for each month
months = tx_retention.columns[2:]
retention_array = [
    {
        'InvoiceYearMonth': months[i + 1],
        'TotalUserCount': np.sum(tx_retention[months[i + 1]]),
        'RetainedUserCount': np.sum(tx_retention.loc[(tx_retention[months[i + 1]] > 0) & (tx_retention[months[i]] > 0), months[i + 1]])
    }
    for i in range(len(months) - 1)
]

tx_retention = pd.DataFrame(retention_array)
tx_retention['RetentionRate'] = tx_retention['RetainedUserCount'] / tx_retention['TotalUserCount']

```

```python
tx_retention.head()
```

```python

# Plot the retention rate graph
plot_data = [
    go.Scatter(
        x=tx_retention.query("InvoiceYearMonth < 201912")['InvoiceYearMonth'],
        y=tx_retention.query("InvoiceYearMonth < 201912")['RetentionRate'],
        mode='lines+markers',
        line=dict(color='red'),
        marker=dict(color='red'),
        hovertemplate='Invoice Year Month: %{x}<br>Retention Rate: %{y}',
        name="organic"
    )
]

plot_layout = go.Layout(
    xaxis=dict(type="category"),
    title='Monthly Retention Rate',
    xaxis_title='Invoice Year Month',
    yaxis_title='Retention Rate',
    paper_bgcolor='white'
)

fig = go.Figure(data=plot_data, layout=plot_layout)
pyoff.iplot(fig)
```

Monthly Retention Rate significantly jumped from June to August and went back to previous levels afterwards.

**Cohort Based Retention Rate**


An alternative method for calculating Retention Rate involves assessing it for each cohort, where cohorts are defined by the first purchase year-month of customers. This approach allows us to measure the percentage of customers retained after their initial purchase in each month. Examining retention rates across cohorts provides insights into the differences between recent and older cohorts, helping us understand whether recent changes in customer experience have impacted the retention of new customers.

```python
#create our retention table again with crosstab() and add firs purchase year month view
tx_retention = pd.crosstab(tx_user_purchase['CustomerID'], tx_user_purchase['InvoiceYearMonth']).reset_index()
tx_retention = pd.merge(tx_retention,tx_min_purchase[['CustomerID','MinPurchaseYearMonth']],on='CustomerID')
new_column_names = [ 'm_' + str(column) for column in tx_retention.columns[:-1]]
new_column_names.append('MinPurchaseYearMonth')
tx_retention.columns = new_column_names
```

```python
#create the array of Retained users for each cohort monthly
retention_array = []
for i in range(len(months)):
    retention_data = {}
    selected_month = months[i]
    prev_months = months[:i]
    next_months = months[i+1:]
    for prev_month in prev_months:
        retention_data[prev_month] = np.nan

    total_user_count = tx_retention[tx_retention.MinPurchaseYearMonth ==  selected_month].MinPurchaseYearMonth.count()
    retention_data['TotalUserCount'] = total_user_count
    retention_data[selected_month] = 1

    query = "MinPurchaseYearMonth == {}".format(selected_month)


    for next_month in next_months:
        new_query = query + " and {} > 0".format(str('m_' + str(next_month)))
        retention_data[next_month] = np.round(tx_retention.query(new_query)['m_' + str(next_month)].sum()/total_user_count,2)
    retention_array.append(retention_data)
```

```python
tx_retention = pd.DataFrame(retention_array)
tx_retention.index = months

#showing new cohort based retention table
tx_retention
```
