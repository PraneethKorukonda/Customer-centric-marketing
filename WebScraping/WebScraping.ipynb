```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://www.newegg.com/p/pl?d={}&page={}"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'
}

def generate_category_url(category):
    return '+'.join(category.split())

def scrape_category(category_name):
    category_url = generate_category_url(category_name)
    all_products = []

    for page_num in range(1, 101):  # Scraping the first 100 pages
        url = base_url.format(category_url, page_num)
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')

        products = soup.find_all('div', class_='item-container')

        for product in products:
            title = product.find('a', class_='item-title').text.strip()
            price_element = product.find('li', class_='price-current')
            if price_element:
                price = price_element.text.strip()
            else:
                price = 'Price not available'

            image = product.find('img')['src']

            product_info = {
                'Title': title,
                'Price': price,
                'Image URL': image
            }
            all_products.append(product_info)


        print(f"Scraped page {page_num} of {category_name}.")

    # Save data to CSV file
    filename = f"{category_name.replace(' ', '')}.csv"
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        fieldnames = ['Title', 'Price', 'Image URL']
        writer = csv.DictWriter(file, fieldnames=fieldnames)

        writer.writeheader()
        writer.writerows(all_products)

    print(f"Scraping for {category_name} finished!")

# Define the list of categories
categories = [
    'Unlocked Phones',
    'Cameras',
    'Camcorders',
    'Coffee Maker',
    'Desktops',
    'Drones',
    'Dslr Cameras',
    'Gaming',
    'Gaming Chairs',
    'Headphones',
    'Home Theater',
    'Keyboards',
    'Laptops',
    'Lenses',
    'Monitors',
    'Mouse',
    'Portable Speakers',
    'Robotic Vacuums',
    'Tablet',
    'Television',
    'Mobiles'
]

# Scraping data for each category
for category_name in categories:
    scrape_category(category_name)
```

```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://www.newegg.com/p/pl?d={}"

categories = [
    'unlocked+phones',
    'cameras',
    'camcorders',
    'CoffeeMaker',
    'Desktops',
    'Drones',
    'DslrCameras',
    'Gaming',
    'GamingChairs',
    'headphones',
    'HomeTheater',
    'Keyboards',
    'Laptops',
    'lenses',
    'Monitors',
    'Mouse',
    'prtablespeakers',
    'RoboticVacuums',
    'TabletData',
    'TelevisionData',
    'Mobiles'
]

def scrape_category(category):
    all_products = []

    for page_num in range(1, 101):  # Scraping the first 100 pages
        url = base_url.format(category) + f"&page={page_num}"
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        products = soup.find_all('div', class_='item-container')

        for product in products:
            title = product.find('a', class_='item-title').text.strip()
            price_element = product.find('li', class_='price-current')
            if price_element:
                price = price_element.text.strip()
            else:
                price = 'Price not available'

            image = product.find('img')['src']

            product_info = {
                'Title': title,
                'Price': price,
                'Image URL': image
            }
            all_products.append(product_info)

        print(f"Category: {category}, Page {page_num} scraped.")

    # Writing data to a CSV file
    filename = f'{category}.csv'
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Title', 'Price', 'Image URL']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for product in all_products:
            writer.writerow(product)

    print(f"Scraping for {category} finished!")

# Scraping data for each category
for category_name in categories:
    scrape_category(category_name)
```
